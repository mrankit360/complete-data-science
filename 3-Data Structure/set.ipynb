```python
s = set(range(1,11))
print(s)
```

```python
set1 = set(range(1,6))
set2 = set(range(1,11,2))

print(f"union: {set1 | set2}")
print(f"intersection: {set1 & set2}")
print(f"difference :{set1 - set2}")
print(f"symmetric difference: {set1 ^ set2}")
```

```python
fs = frozenset(range(1,6)) #FROZEN SET IS IMMUTABLE
print(fs)
```

```python
s = set(range(1,6))
print(s)
s.pop()
print(s)
print(3 in s)
```
