### File Handling And Operation

```python
def read_file(filename):
    with open(filename, 'r') as file:
        for line in file:
            print(line.strip())

read_file('sample.txt')
```

```python
def write_file(lines, filename):
    with open(filename, 'w') as file:
        for line in lines:
            file.write(line + '\n')

write_file(['hello', 'ankit how are you'], 'output.txt')            
```

```python
def count_words(filename):
    with open(filename, 'r') as file:
        text = file.read()
        words = text.split()
        return len(words)

count_words('output.txt')
```

```python
def append_to_file(text, filename):
    with open(filename, 'a') as file:
        file.write(text + '\n')

append_to_file('what are you doing', 'output.txt')        
```

```python
def read_reverse(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    for line in reversed(lines):
        print(line.strip())

read_reverse('sample.txt')        
```
