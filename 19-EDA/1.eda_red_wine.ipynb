## EDA WITH RED WINE DATASET

 The Red Wine Quality dataset is widely used for analyzing and predicting wine quality based on its chemical properties. This dataset contains various features that influence wine quality, such as acidity, sugar levels, alcohol content, and more. Below is a summary of the dataset and key results derived from its analysis.

1) Alcohol
2) Malic acid
3) Ash
4) Alcalinity of ash  
5) Magnesium
6) Total phenols
7) Flavanoids
8) Nonflavanoid phenols
9) Proanthocyanins
10) Color intensity
11) Hue
12) OD280/OD315 of diluted wines
13) Proline 

Output variable (based on sensory data)

:- quality (score between 0 and 10 )
