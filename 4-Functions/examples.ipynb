```python
## TEMPERATURE CONVERSION

def convert_temperature(temp, unit):
    """This function convert temp"""
    if unit == 'C':
        return temp * 9/5 + 32
    elif unit == 'F':
        return (temp-32)*5/9
    else:
        return None
    
print(convert_temperature(25, 'C'))
print(convert_temperature(77, 'F'))

```

```python
def calculate_cost(cart):
    total_cost = 0
    for item in cart:
        total_cost+= item['price']*item['quantity']

    return total_cost

cart = [
    {'name':'apple', 'price':5, 'quantity':4},
    {'name':'cherry', 'price':10, 'quantity':2},
    {'name':'litchi', 'price':7, 'quantity':4}
]

calculate_cost(cart)
```

```python
def is_pallindrome(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]

print(is_pallindrome("A man a plan A canal Panama"))
```
