```python
def even(num):
    if num%2 == 0:
        return True

even(18)
```

```python
even1 = lambda num:num%2 == 0
even1(24)
```

```python
## MAP function :- apply function to all item in list, tuple
numbers = [1,2,3,4,5]
def square(number):
    return number**2

square(2)
```

```python
list(map(square, numbers))
```

```python
list(map(lambda x:x**2, numbers))
```

```python
number1 = [1,2,3]
number2 = [4,5,6]

list(map(lambda x,y:x+y, number1,number2))
```

```python
words = ['apple', 'banana', 'cherry']
upper_word = list(map(str.upper, words))
print(upper_word)

list(map(len, words))
```

```python
## FILTER FUNCTION
numbers = [1,2,3,4,5,6]
list(filter(lambda x:x%2==0, numbers))

```
