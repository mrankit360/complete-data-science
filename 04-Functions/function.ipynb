```python
def even_odd(num):
    """This function finds even or odd"""
    if num%2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")

even_odd(23)
```

```python
def greet(name="guest"):
    print(f"{name} welcome to the club")

greet()
```

```python
def print_number(*ankit): ## *args is variable length argument
    for number in ankit:
        print(number)

print_number(1,2,3,4,5,6,7,"ankit")
```

```python
def print_details(**kwargs):  # **kwargs store key, value data
    for key, value in kwargs.items():
        print(f"{key} : {value}")

print_details(name="ankit", age=20, country="india")
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f"positional argument: {val}")

    for key, value in kwargs.items():
        print(f"{key} : {value}")

print_details(1,2,3,4,"ankit",name="ankit",age=20,country="india") 
```
