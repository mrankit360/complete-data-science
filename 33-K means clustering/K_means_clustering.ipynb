## K Means Clustering Implementations

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
X,y=make_blobs(n_samples=1000,centers=3,n_features=2)
```

```python
X,y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
## Standardization -- feature scaling technique
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42) ## we ignore dependent feature y
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.cluster import KMeans
```

```python
## Elbow method to select k value
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
##plot elbow curve
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel('Number of cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++')
```

```python
kmeans.fit_predict(X_train_scaled)
```

```python
y_pred=kmeans.predict(X_test_scaled)
```

```python
y_pred
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=y_pred)
```

```python
## validating the k value
## kneelocator
## silhoutee scoring
```

```python
## Kneelocator pip install kneed
from kneed import KneeLocator
```

```python
kl=KneeLocator(range(1,11),wcss,curve='convex',direction='decreasing')
```

```python
kl.elbow
```

```python
## silhoutte scoring
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(X_train_scaled)
    score=silhouette_score(X_train_scaled,kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
##plot elbow curve
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel('Number of cluster')
plt.ylabel('silhoutte coeficients')
plt.show()
```
