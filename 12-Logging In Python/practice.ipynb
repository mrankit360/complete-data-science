```python
import logging

logging.basicConfig(
    filename='app1.log',
    filemode='a',
    level=logging.WARNING,
    format= '%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%y-%m-%d %h:%m:%s'
    )

def login(username,password):
    original_user = 'ankit'
    original_pass = '1234'

    if(original_user == username and original_pass == password):
        logging.info('you can login')
        return('login successful')

    else:
        logging.warning('wrong username or password')
        return('invalid username or password')

print(login('ankita', '1234'))
```
