```python
## Password Checker
import logging

logging.basicConfig(
    filename="security.log",
    filemode='a',
    level=logging.WARNING, #only warning level will be saved in file
    format="%(asctime)s-%(levelname)s-%(message)s",
    datefmt="%y-%m-%d %H:%M:%S",
    #force=True
)

def login(username,password):
    correct_username = "Ankit"
    correct_password = "12345"

    if username == correct_username and password==correct_password:
        logging.info(f"User '{username}' logged in successfully.")
        return "login successfull"
    else:
        logging.warning(f"User '{username}' and Password '{password}' is wrong")
        return "Invalid username and password"

print(login("ankit","1234"))
#print(login("Ankit","12345"))
```
