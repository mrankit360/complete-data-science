```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df.head(5)
```

```python
# Handling Missing values
df.isnull().any()
df.isnull().sum()
```

```python
df_filled = df.fillna(0)
df_filled.isnull().sum()
```

```python
# Missing value with mean of column
df['sales_fill'] = df['Sales'].fillna(df['Sales'].mean())
df['sales_fill'].isnull().any()
df.head(3)
```

```python
# renaming column
df = df.rename(columns={'Date':'Sales_date'})
df.head(2)
```

```python
#Change data type
df['Value_new'] = df['Value'].fillna(df['Value'].mean()).astype(int)
df.head(3)
```

```python
df['New_value'] = df['Value'].apply(lambda x:x**3)
df.head(3)
```

```python
# Data aggregating and grouping

df.head()
grouped_mean = df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(['Product','Region'])['Value'].sum()
grouped_sum
```

```python
## aggregate multiple function
group_agg = df.groupby('Region')['Value'].agg(['mean','sum','count'])
group_agg
```

```python
### Merging and joining Dataframes
# Create sample DataFrames
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})
```

```python
df1
```

```python
##Merging dataframe
pd.merge(df1, df2, on="Key",how="inner")
```

```python
pd.merge(df1, df2, on="Key",how="outer")
```

```python
pd.merge(df1,df2,on="Key",how="left")
```

```python
pd.merge(df1,df2,on="Key",how="right")
```
