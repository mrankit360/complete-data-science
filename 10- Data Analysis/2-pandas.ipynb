## Pandas
Data manipulation library widely used for data analysis and cleaning.

Types of data structure
1. series :- one dimension Array
2. Dataframe :- two dimension array

```python
import pandas as pd
```

```python
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
```

```python
# create series from dictionary

data ={'a':1, 'b':2, 'c':3}
series_dic = pd.Series(data)
print(series_dic)

data = [10,20,30]
index = ['a','b','c']
print(pd.Series(data, index))
```

```python
# Data Frame
data = {
    'Name' : ['Krish', 'Ankit'],
    'Age' : [30, 20],
    'city' : ['mumbai', 'jabalpur']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
# Accessing data from dataframe
data['Name']
type(data['Name'])
```
