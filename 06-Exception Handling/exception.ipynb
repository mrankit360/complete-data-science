### Exception Handling with try,except,finally

```python
## Zero Division

def divide(a,b):
    try:
        result = a / b
    except ZeroDivisionError as e:
        print(f"Error: {e}")
        result = None
    finally:
        print("execution success")
    return result

divide(10,2)
divide(10,0)
```

```python
def read_file(filename):
    try:
        file = open(filename, 'r')
        content = file.read()
        return content
    except FileNotFoundError as e:
        print(f"Error: {e}")
    finally:
        try:
            file.close()
        except NameError:
            pass

read_file('outut.txt')        
```

```python
def get_integer():
    try:
        value = int(input("Enter an integer: "))
    except ValueError as e:
        print(f"Error: {e}")
        value = None
    finally:
        print("Execution complete.")
    return value

get_integer()
```
