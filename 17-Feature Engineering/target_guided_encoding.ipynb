```python
import pandas as pd

#create sample dataframe with categorical variable and a target variable

df=pd.DataFrame({
    'city': ['New York', 'London', 'Paris', 'Tokyo', 'New York', 'Paris'],
    'price': [100,150,300,250,180,320]
})
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)
```

```python
df[['price','city_encoded']]
```
